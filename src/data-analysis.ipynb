## Análise dos dados

Criando para efetuar análise dos dados, visando entender melhor o data-set.

```python
import pandas as pd

from pymongo import MongoClient
import matplotlib.pyplot as plt

# Display plots inline and change default figure size
%matplotlib inline
```

```python
client = MongoClient()
client = MongoClient('localhost', 27800)

db = client.network_logs

db_un_hostnames = db.unique_hostnames
db_un_urls = db.unique_host_urls

db_odp = client.odp_logs 
db_log_lines = db_odp.log_lines
```

```python
df_hostnames = pd.DataFrame(list(db_un_hostnames.find({})))
df_urls = pd.DataFrame(list(db_un_urls.find({})))
df_odp_urls = pd.DataFrame(list(db_log_lines.find({})))
```

```python
pd.set_option('display.max_rows', None)
print(df_hostnames['category'].nunique())
print(len(df_hostnames))
df_hostnames.groupby('category').count()
```

```python
pd.set_option('display.max_rows', None)
print(df_urls['category'].nunique())
print(len(df_urls))
df_urls.groupby('category').count()
```

```python
plt.figure(1,figsize=(25,10))
df_hostnames.groupby('category')['normalized_hostname'].count().plot.bar()
```

```python
plt.figure(1,figsize=(25,10))
df_urls.groupby('category')['normalized_hostname_url'].count().plot.bar()
```

```python
pd.set_option('display.max_rows', None)
print(df_odp_urls['category'].nunique())
print(len(df_odp_urls))
df_odp_urls.groupby('category').count()
```

```python
plt.figure(1,figsize=(25,10))
df_odp_urls.groupby('category')['normalized_url'].count().plot.bar()
```
